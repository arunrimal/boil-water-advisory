##### **For the puurpose of NER, we need to create a custome entity that will we labled as 'Advisory Level'. For that purpose all the advisory reasons were extractes fro the BWA dataset.**

```python
import pandas as pd
import json
import re

# Loading the Excel file
df = pd.read_excel("Data/BWA_KS.xlsx")

# Dropping the missing values and getting unique reasons
unique_reasons = df["Reason"].dropna().unique()
```

```python
unique_reasons[0:10]
```

##### **Now cleaning the extracted list**

```python
# converting to lowercase and remove punctuation and extra spaces
# remove punctuation by matching anything that is NOT a letter, number, 
# underscore, or space

def clean_reason(text):
    text = text.lower()                     
    text = text.strip()                     
    text = re.sub(r"[^\w\s]", "", text)     
    text = re.sub(r"\s+", " ", text)        
    return text
```

```python
cleaned_reasons = [clean_reason(reason) for reason in unique_reasons]
cleaned_reasons[0:10], len(cleaned_reasons)
```

```python
cleaned_reasons = sorted(set(cleaned_reasons))
cleaned_reasons[0:10], len(cleaned_reasons)
```

```python
reason_dictionary = {
    reason: reason.split()   # creating a keyword list
    for reason in cleaned_reasons
}
reason_dictionary
```

```python
with open("reason_dictionary.json", "w") as f:
    json.dump(reason_dictionary, f, indent=4)
```

##### **Merging all 8 json files into one for further processing like self joining and cross joining later in the data preparation phase**
 

```python
import json
import os
import glob

input_folder = "advisory_reasons_output" 
output_folder = "merged_json_output"

output_file = os.path.join(output_folder, "merged_output.json")

os.makedirs(output_folder, exist_ok=True)

files = glob.glob(os.path.join(input_folder, "*.json"))
merged = []

for filepath in sorted(files):
    with open(filepath, "r", encoding="utf-8") as f:
        data = json.load(f)
    if isinstance(data, list):
        merged.extend(data)
    else:
        merged.append(data)
    print(f" Loaded: {os.path.basename(filepath)}")

with open(output_file, "w", encoding="utf-8") as f:
    json.dump(merged, f, indent=2, ensure_ascii=False)

print(f"\nDone! Merged {len(files)} files -> '{output_file}' ({len(merged)} total records)")
```
